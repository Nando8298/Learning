Import Class

```python
import pandas as pd
import numpy as np
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import cross_val_score
```

***Read dataset from where***

```python
data = pd.read_csv("shrimp_biomass_cleaned(1).csv")
data.head()
```

Dropping (x):
- Mortality rate per day (Target(y))
- Cumulative Mortality
- Ancho pakan 1,2,3
- Standard Deviation
- Pakan Difference, %
Other than the target, the variables listed above do not affect the result.

Using OneHotEncoder to change WA(Weather) from object to int64
ColumnTransformer for replacing old WA(Weather) values with new values
Remaining columns are kept unchanged and appended to the transformed output 

```python
x = data.drop(["M_Hr", "MOR_Kum", "Pakan_Ancho_1", "Pakan_Ancho_2", "Pakan_Ancho_3", "Std", "Pakan_diff", "Pakan_diff_%"], axis = 1)
y = data["M_Hr"]

categorical_features = ["WA_6", "WA_14"]
ohe = OneHotEncoder()
transform = ColumnTransformer([("one_hot", ohe, categorical_features)], remainder = "passthrough")
transformed = transform.fit_transform(x)
```

Splitting the dataset into train_test_split with 20% test size

```python
x_train, x_test, y_train, y_test = train_test_split(transformed,y,test_size= 0.20)
```

Using RandomForestRegressor for first model

```python
rfr = RandomForestRegressor()
rfr.fit(X= x_train, y= y_train)
```

```python
y_pred = rfr.predict(x_test)
```

```python
score = rfr.score(x_test, y_test)
score
```

With RandomForestRegressor, the score is 47.4%

Trying cross_val_score to cross evaluate 5 models with different estimators.

```python
for i in range(100,200,10):
   print(i, " estimators")
   model = RandomForestRegressor(n_estimators=i)
   model.fit(x_train, y_train)

   model_score = model.score(x_test, y_test)
   print(model_score)

   mean = np.mean(cross_val_score(model, transformed, y, cv = 5))
   print(mean)
   print("")
    
```
